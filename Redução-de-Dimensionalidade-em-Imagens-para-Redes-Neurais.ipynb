```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Carrega a imagem
img1 = mpimg.imread('teste/6.jpg')
```

```python
# Armazena altura, largura e tambem os valores binários para a img_cinza e img_binaria
altura, largura, _ = img1.shape
img_cinza = [[0 for _ in range(largura)] for _ in range(altura)]
img_binaria = [[0 for _ in range(largura)] for _ in range(altura)]
```

```python
# Converter para escala de cinza
for i in range(altura):
    for j in range(largura):
        r, g, b = img1[i][j]
        cinza = int(0.299 * r * 255 + 0.587 * g * 255 + 0.114 * b * 255)  # cinza
        img_cinza[i][j] = cinza / 255.0  # Normalizar para o intervalo [0, 1]
```

```python
# Definir um limiar para a imagem binária
limiar = 120

# Converter de cinza para binário
for i in range(altura):
    for j in range(largura):
        img_binaria[i][j] = 1 if img_cinza[i][j] > limiar else 0
```

```python
# Imagem
plt.subplot(1, 3, 1)
plt.imshow(img1)
plt.title('Imagem')
plt.axis('off')
plt.show()
```

```python
# Imagem em escala de cinza
plt.subplot(1, 3, 2)
plt.imshow(img_cinza, cmap='gray')
plt.title('Imagem em escala de cinza')
plt.axis('off')
plt.show()
```

```python
# Imagem binária
plt.subplot(1, 3, 3)
plt.imshow(img_binaria, cmap='gray')
plt.title('Imagem Binária')
plt.axis('off')
plt.show()
```
